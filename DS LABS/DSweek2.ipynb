```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Age': [25, 30, np.nan, 40, 35],
    'Department': ['HR', 'Finance', 'Finance', np.nan, 'IT']
})

df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Department'].fillna(df['Department'].mode()[0], inplace=True)

print(df)
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Age': [25, 30, np.nan, 40, 35],
    'Department': ['HR', 'Finance', 'Finance', np.nan, 'IT']
})

df_ffill = df.copy()
df_ffill.fillna(method='ffill', inplace=True)

print(df_ffill)
```

```python
import pandas as pd

df = pd.DataFrame({
    'ID': [1, 2, 2, 3, 4, 4],
    'Name': ['Alice', 'Bob', 'Bob', 'Charlie', 'David', 'David'],
    'Age': [25, 30, 30, 35, 40, 40]
})

print("Original Data:\n", df)

df_exact = df.drop_duplicates()
print("\nAfter Exact Match Removal:\n", df_exact)
```

```python
import pandas as pd

df = pd.DataFrame({
    'ID': [1, 2, 2, 3, 4, 4],
    'Name': ['Alice', 'Bob', 'Bob', 'Charlie', 'David', 'David'],
    'Age': [25, 30, 30, 35, 40, 40]
})

df_subset_id = df.drop_duplicates(subset=['ID'])
print("After Subset-Based Removal (ID):\n", df_subset_id)

df_subset_name = df.drop_duplicates(subset=['Name'])
print("After Subset-Based Removal (Name):\n", df_subset_name)
```

```python
import pandas as pd

df = pd.DataFrame({
    'Name': ['Alice', 'BOB', 'charlie', 'DAVID']
})

df['Name_lower'] = df['Name'].str.lower()
df['Name_upper'] = df['Name'].str.upper()

print(df)
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({'Values': [10, 12, 14, 15, 13, 100, 11, 14, 13, 200]})

Q1 = df['Values'].quantile(0.25)
Q3 = df['Values'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['Values'] < lower_bound) | (df['Values'] > upper_bound)]
print("Outliers:\n", outliers)
```

```python
import pandas as pd
import numpy as np
from scipy import stats

df = pd.DataFrame({'Values': [10, 12, 14, 15, 13, 100, 11, 14, 13, 200]})

z_scores = np.abs(stats.zscore(df['Values']))
outliers = df[z_scores > 3]

print("Outliers:\n", outliers)
```
