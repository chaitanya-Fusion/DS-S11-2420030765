```python
# Mean
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()

df = pd.DataFrame(
    iris.data,
    columns=iris.feature_names
)

print("First 5 rows of dataset:")
print(df.head())

mean_values = df.mean()

print("\nMean values:")
print(mean_values)
```

```python
# Median
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()

df = pd.DataFrame(
    iris.data,
    columns=iris.feature_names
)

print("First 5 rows of dataset:")
print(df.head())

median_values = df.median()

print("\nMedian values:")
print(median_values)
```

```python
# Mode
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()

df = pd.DataFrame(
    iris.data,
    columns=iris.feature_names
)

print("First 5 rows of dataset:")
print(df.head())

mode_values = df.mode().iloc[0]

print("\nMode values:")
print(mode_values)
```

```python
# Range
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()

df = pd.DataFrame(
    iris.data,
    columns=iris.feature_names
)

print("First 5 rows of dataset:")
print(df.head())

range_values = df.max() - df.min()

print("\nRange values:")
print(range_values)
```

```python
# Variance
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()

df = pd.DataFrame(
    iris.data,
    columns=iris.feature_names
)

print("First 5 rows of dataset:")
print(df.head())

variance_values = df.var()

print("\nVariance values:")
print(variance_values)
```

```python
# Standard Deviation
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()

df = pd.DataFrame(
    iris.data,
    columns=iris.feature_names
)

print("First 5 rows of dataset:")
print(df.head())

std_values = df.std()

print("\nStandard Deviation values:")
print(std_values)
```

```python
# Interquartile Range (IQR)
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()

df = pd.DataFrame(
    iris.data,
    columns=iris.feature_names
)

print("First 5 rows of dataset:")
print(df.head())

Q1 = df.quantile(0.25)

Q3 = df.quantile(0.75)

IQR = Q3 - Q1

print("\nInterquartile Range (IQR):")
print(IQR)
```

```python
# Skewness
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()

df = pd.DataFrame(
    iris.data,
    columns=iris.feature_names
)

print("First 5 rows of dataset:")
print(df.head())

skewness_values = df.skew()

print("\nSkewness values:")
print(skewness_values)
```

```python
# Kurtosis
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()

df = pd.DataFrame(
    iris.data,
    columns=iris.feature_names
)

print("First 5 rows of dataset:")
print(df.head())

kurtosis_values = df.kurt()

print("\nKurtosis values:")
print(kurtosis_values)
```
